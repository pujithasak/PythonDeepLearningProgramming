```python
"""

Apply  the  three  classification  algorithms  Na√Øve Bayes,  SVM  and  KNN  on  the chosen data set and
report which classifier gives better result.

"""

# importing libraries to python
from sklearn import model_selection
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
import pandas as pd
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

# reading the csv file data using pandas library
wine_data = pd.read_csv('C:/Users/anish/PycharmProjects/PythonCourse/PythonProject/wine_dataset.csv')

# Check the null values
nulls = pd.DataFrame(wine_data.isnull().sum().sort_values(ascending=False))
nulls.columns = ['Null Count']
nulls.index.name = 'Feature'

# Remove the null values
new_data = wine_data.dropna()

# preprocessing the data
x = new_data.drop('Wine Type', axis=1)
y = new_data['Wine Type']

# splitting data into training data and testing data
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(x, y, test_size=0.2, random_state=0)

# creating and Training the Naive Bayes classifier from sklearn library
gb = GaussianNB()
gb.fit(X_train, Y_train)

# Predicting the data
Y_pred = gb.predict(X_test)

# Display the Score
print("\n Gaussian Naive Bayes Accuracy Score:", metrics.accuracy_score(Y_test, Y_pred))
print("\n Classification Report: \n", metrics.classification_report(Y_test, Y_pred))


# creating and training the SVM classifier from sklearn library
svc = SVC(kernel='linear')
# Predicting the data
Y_pred = svc.fit(X_train, Y_train).predict(X_test)

# Display the score
print("\n Linear SVM Accuracy Score:", metrics.accuracy_score(Y_test, Y_pred), "\n")
print("\n Classification Report: \n", metrics.classification_report(Y_test, Y_pred))

# creating and training the KNN classifier from sklearn library
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, Y_train)

# Predicting the data
Y_pred = knn.predict(X_test)

# Display the Score
print("\n KNN Accuracy Score:", metrics.accuracy_score(Y_test, Y_pred))
print("\n Classification Report: \n", metrics.classification_report(Y_test, Y_pred))
```
